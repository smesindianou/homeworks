```python
import pandas as pd
import numpy as np
```

```python
av_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/av/cleanav_articles.csv')
```

```python
kath_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/kathimerini/clean_articles_kath.csv')
```

```python
naft_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/naftemporiki/clean_articles_naft.csv')
```

```python
tt_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/the toc/cleaned_articles_tt.csv')
```

```python
vima_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/vima/cleaned_articles_vima.csv')
```

```python
g_trends_data = pd.read_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/multiTimeline.csv', skiprows=2)
```

```python
av_data['Μέσο'] = 'Athens Voice'
```

```python
kath_data['Μέσο'] = 'Καθημερινή'
```

```python
naft_data['Μέσο'] = 'Ναυτεμπορική'
```

```python
tt_data['Μέσο'] = 'The Toc'
```

```python
vima_data['Μέσο'] = 'Το ΒΗΜΑ'
```

```python
vima_data = vima_data.rename(columns={'excert': 'excerpt'})
```

```python
all_mme_data = pd.concat([av_data, kath_data, naft_data, tt_data, vima_data], ignore_index=True)
all_mme_data.shape
```

```python
all_mme_data.columns
```

```python
all_mme_data['date'] = pd.to_datetime(all_mme_data['date'], format = '%Y-%m-%d')
```

```python
all_mme_data.dtypes
```

```python
all_mme_data_short = all_mme_data[['title', 'section', 'author', 'excerpt', 'text_body', 'date', 'Μέσο']]
```

```python
all_mme_data_short = all_mme_data_short[~all_mme_data_short['text_body'].isna()].reset_index(drop=True)
```

```python
all_mme_data_short.sort_values("date", ascending=False)
```

```python
all_mme_data_short['text'] = all_mme_data_short[['title', 'excerpt', 'text_body']].fillna('').agg(' '.join, axis=1)
```

```python
all_mme_data_short = all_mme_data_short.drop(columns=['title', 'excerpt', 'text_body'])
```

```python
all_mme_data_short = all_mme_data_short.sort_values("date", ascending=False).reset_index(drop=True)
```

```python
all_mme_data_short
```

```python
all_mme_data_short = all_mme_data_short[['date', 'text', 'section', 'Μέσο']]
```

```python
all_mme_data_short = all_mme_data_short.sort_values("date", ascending=False).reset_index(drop=True)
```

```python
all_mme_data_short.to_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/all/all_mme_data.csv', index = False)
```
