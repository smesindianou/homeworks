```python
g_baddf_av2 = g_baddf_av[~g_baddf_av['section'].isin(['Κόσμος', ])]
g_baddf_av2.shape
```

```python
# κοιτάω με το unique() να δω αν οι στήλες date και time έχουν μόνο NaN values
print(f"date unique values: {g_baddf_av2['date'].unique()}")
print(f"time unique values: {g_baddf_av2['time'].unique()}")
```

```python
# Μετατροπή σε datetime με συγκεκριμένη μορφή
g_baddf_av2['date'] = pd.to_datetime(g_baddf_av2['date'], format='%d-%m-%Y')
```

```python
g_baddf_av2.sort_values("date")
```

```python
g_baddf_av2.sort_values("date", ascending=False).reset_index(drop=True)
```

Και για να σωθεί το σορτάρισμα και το reset του index...

```python
g_baddf_av2 = g_baddf_av2.sort_values("date", ascending=False).reset_index(drop=True)
```

```python
g_baddf_av2['year'] = g_baddf_av2['date'].dt.year
g_baddf_av2['month'] = g_baddf_av2['date'].dt.month
g_baddf_av2['day'] = g_baddf_av2['date'].dt.day
g_baddf_av2
```

```python
g_baddf_av2.to_csv('/content/drive/MyDrive/ΓΥΝΑΙΚΟΚΤΟΝΙΑ_Ε_P/vima/cleaned_articles_vima.csv', index = False)
```
